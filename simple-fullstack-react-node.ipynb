# Simple Full Stack App: React + Node.js
This notebook guides you through building a simple full stack app using React for the frontend and Node.js/Express for the backend. The UI is interactive and easy to follow.

## 1. Set Up Node.js Backend
We'll start by initializing a Node.js project, installing Express, and creating a simple API endpoint.

**Terminal Commands:**
```bash
# In your project folder
mkdir backend
cd backend
npm init -y
npm install express cors
```

**Create backend/server.js:**

````python
// backend/server.js
const express = require('express');
const cors = require('cors');
const app = express();
const PORT = 5000;

app.use(cors());

app.get('/api/message', (req, res) => {
  res.json({ message: 'Hello from Node.js backend!' });
});

app.listen(PORT, () => {
  console.log(`Server running on http://localhost:${PORT}`);
});
````

---
## 2. Create React Frontend
Set up a React app and prepare the main App component.

**Terminal Commands:**
```bash
# In your project folder (not inside backend)
npx create-react-app frontend
cd frontend
npm start
```

**Edit frontend/src/App.js:**

````python
// frontend/src/App.js
import React, { useState } from 'react';

function App() {
  const [message, setMessage] = useState('');

  return (
    <div style={{ textAlign: 'center', marginTop: 50 }}>
      <h1>Simple React + Node.js App</h1>
      <button>Get Message from Backend</button>
      <p>{message}</p>
    </div>
  );
}

export default App;
````

---
## 3. Connect React to Node.js Backend
Update the React button to fetch data from the backend API and display it.

````python
// frontend/src/App.js (update the button handler)
import React, { useState } from 'react';

function App() {
  const [message, setMessage] = useState('');

  const getMessage = async () => {
    const res = await fetch('http://localhost:5000/api/message');
    const data = await res.json();
    setMessage(data.message);
  };

  return (
    <div style={{ textAlign: 'center', marginTop: 50 }}>
      <h1>Simple React + Node.js App</h1>
      <button onClick={getMessage}>Get Message from Backend</button>
      <p>{message}</p>
    </div>
  );
}

export default App;
````

---
## 4. Add Simple Interactive UI in React
The button now fetches and displays a message from the backend. You can further style or expand the UI as needed.
